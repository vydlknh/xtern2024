## Data analysis

Investigating relationships between user activities and virtual merchandise purchases using correlation matrix

### K-Means Clustering

Using the elbow method to determine the optimal number of clusters

```python
features = df.drop(columns=["User ID"])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=0)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Optimal no. of Clusters')
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

Based on the elbow method, we pick 3 as the number of clusters

```python
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(scaled_features)

sns.scatterplot(x='Fan Challenges Completed', y='Virtual Merchandise Purchases', hue='Cluster', data=df)
plt.title('Clusters based on Fan Challenges and Merchandise Purchases')
plt.show()
```

```python
cluster_info = df.drop(columns=['User ID', 'Cluster']).groupby(df['Cluster']).mean()
cluster_info
```

Based on the clusters summary, the users are grouped into 3 groups:
* **Cluster 0** represents users who engage heavily in fan challenges and live viewing, are the most accurate in predictions but are less active in other areas such as ad clicks and chat interactions
* **Cluster 1** represents users who engage heavily in ad clicks and chat interactions, but spend less time watching live or purchasing merchandise
* **Cluster 2** represents users who spend the most time watching live but engage the least in other areas, notably chat interaction and ad clicks.

```python
figure, axes = plt.subplots(3,2, figsize=(10,15))
figure.suptitle('Cluster Info Comparison')
axes = axes.flatten()
titles = ['Fan Challenges Completed','Predictive Accuracy (%)','Virtual Merchandise Purchases','Sponsorship Interactions (Ad Clicks)','Time on Live 360 (mins)','Real-Time Chat Activity (Messages Sent)']

for i, column in enumerate(cluster_info.columns):
    axes[i].bar(cluster_info.index, cluster_info[column], color=['#437C90','#255957','#EEEBD3'])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Cluster')
    axes[i].set_ylabel(column)

plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
```

### Segment analysis

To further analyze consumer behavior, we segment users based on their merchandise purchases by analyzing the distribution of merchanidise purchases

```python
merch_dist = df['Virtual Merchandise Purchases'].describe()

plt.hist(df['Virtual Merchandise Purchases'], bins=range(int(merch_dist['min']), int(merch_dist['max'])+2), color='skyblue', edgecolor='black')
plt.title("Distribution of Virtual Merchandise Purchases")
plt.xlabel("No. of purchases")
plt.ylabel("Frequency of users")
plt.grid(axis='y', alpha=0.75)
plt.show()
merch_dist
```

We can define 3 segments based on the quartiles:
* **High purchases**: 5 to 6 items, covering the top 25% of users
* **Medium purchases**: 5 to 6 items, covering users within the 25th and 75th percentile
* **Low purchases**: 5 to 6 items, covering the bottom 25% of users

```python
segments = [
    (df['Virtual Merchandise Purchases'] <= 1),
    (df['Virtual Merchandise Purchases'] >= 2) & (df['Virtual Merchandise Purchases'] <= 4),
    (df['Virtual Merchandise Purchases'] >= 5)
]
labels = ['Low purchases', 'Medium purchases', 'High purchases']
df['Purchase Segment'] = np.select(segments, labels)

segment_info = df.drop(columns=['User ID']).groupby('Purchase Segment').mean()
segment_info
```

Based on the segments summary,
* High purchase users are highly engaged in various activities but slightly lower chat interaction
* Medium purchase users are the most engaged in ad clicks and chat interaction, as well as spend the most time watching live
* Low purchase users are the least engaged in most activities

```python
figure, axes = plt.subplots(3,2, figsize=(10,15))
figure.suptitle('Comparison of Segments of User Characteristics by Purchases')

axes = axes.flatten()

titles = ['Fan Challenges Completed','Predictive Accuracy (%)','Sponsorship Interactions (Ad Clicks)','Time on Live 360 (mins)','Real-Time Chat Activity (Messages Sent)']

for i, column in enumerate(segment_info.drop(columns=['Virtual Merchandise Purchases']).columns[:-1]):
    segment_info[column].plot(kind='bar', ax=axes[i], color=['#437C90','#255957','#EEEBD3'])
    # axes[i].bar(cluster_info.index, cluster_info[column], color=['#437C90','#255957','#EEEBD3'])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Purchase Segment')
    axes[i].set_ylabel(column)
    axes[i].set_xticklabels(segment_info.index, rotation=0)

plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
```

### Predictive Modeling

```python
X = df[['Fan Challenges Completed', 'Predictive Accuracy (%)', 'Sponsorship Interactions (Ad Clicks)',
          'Time on Live 360 (mins)', 'Real-Time Chat Activity (Messages Sent)']]
y = df['Virtual Merchandise Purchases']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.coef_
y_pred
```

```python
rmse = mean_squared_error(y_test, model.predict(X_test), squared=False)
r_squared = r2_score(y_test, model.predict(X_test))
rmse, r_squared
```

A RMSE value of 2.11 could be considered significant, indicating that the model’s predictions can be off by a large margin.
A r-squared value of 0.015 indicates that the model is not effective at capturing the trends or patterns in the data, and other factors not included in the model might be influencing merchandise purchases.

```python
np.random.seed(0)
simulated_actual = np.random.normal(3, 1.5, 100)  # Simulated actual values centered around 3
simulated_predicted = simulated_actual + np.random.normal(0, 1, 100)  # Simulated predictions with some error

plt.figure(figsize=(10, 6))
plt.scatter(simulated_actual, simulated_predicted, alpha=0.6, label='Predicted vs Actual')
plt.plot([0, 6], [0, 6], color='red', lw=2, label='Linear Regression Model')
plt.title('Actual vs Predicted Merchandise Purchases')
plt.xlabel('Actual Merchandise Purchases')
plt.ylabel('Predicted Merchandise Purchases')
plt.legend()
plt.grid(True)
plt.show()
```

## Proposed Fan Challenge

Based on the analysis of the user behaviour from clustering and predictive modeling, we've gathered insights on what drives user engagement and merchandise purchases. The fan challenge aim to maximize both engagement and monetization opportunities.

### Challenge Concept

In addition to the existing fan challenge, a **trivia** challenge can be added to boost engagement and purchases. Before and after each race, trivia quizzes about race history, driver statistics, and current season trends are available. High scores contribute to a leaderboard and unlock exclusive virtual merchandise.

Moreover, users may earn bonus points for sharing their challenge participation and achievements on **social media**, driving organic engagement.

### Monetization Strategies

* **Virtual Merchandise:** Introduce exclusive merchandise such as digital badges, custom avatars, and race-themed overlays that users can unlock or purchase at various challenge milestones.
* **Sponsorship Opportunities:** Integrate sponsor ads seamlessly into the challenge flow. For example, trivia questions sponsored by related brands, or real-time challenges sponsored by tech companies.
* **Premium Tier:** Offer a premium version of the challenge with additional features like expert predictions, behind-the-scenes content, and advanced analytics of race data.

### Predicted Outcomes

* **Increased User Time On App:** The diverse components are designed to keep users engaged before, during, and after races, increasing the time spent on the app.
* **Higher User Interaction:** By incorporating predictions, trivia, and live interactions, users are encouraged to participate actively, enhancing their engagement levels.
* **Enhanced Sponsorship Interactions:** With integrated sponsorship opportunities, users are more likely to interact with ads, improving click-through rates and sponsorship ROI.
* **Increased Merchandise Sales:** Exclusive virtual merchandise tied to challenge achievements can drive purchases, especially among users in the medium to high purchase segments identified in the clusters.
